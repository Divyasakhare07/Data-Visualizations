```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Combined.xlsx', sheet_name='Pasco')
```

```python
data.columns = ['Date', 'Daily Cases', '7-Day Moving Average']
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%b-%y')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Daily Cases'], label='Daily Cases', color='blue')
plt.plot(data['Date'], data['7-Day Moving Average'], label='7-Day Moving Average', color='orange')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Daily Cases and 7-Day Moving Average in Pasco')
plt.legend()
plt.show()
```
