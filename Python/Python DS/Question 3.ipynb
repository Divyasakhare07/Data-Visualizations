```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Combined.xlsx', sheet_name='Hillsborough')
```

```python
data.columns = ['Date', 'Daily Cases', '7-Day Moving Average']
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%B %d, %Y')
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Date'], data['7-Day Moving Average'], color='skyblue', label='7-Day Moving Average')
plt.plot(data['Date'], data['Daily Cases'], color='darkblue', label='Daily Cases', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Daily Cases and 7-Day Moving Average in Hillsborough')
plt.legend()
plt.show()
```
