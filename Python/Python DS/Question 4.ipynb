```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Car Sales.xlsx')
```

```python
data['TotalCost'] = data['CostPrice'] + data['LaborCost']
```

```python
country_summary = data.groupby('CountryName').agg({'SalePrice': 'sum', 'TotalCost': 'sum'}).reset_index()
```

```python
country_summary = country_summary[country_summary['CountryName'].isin(['United Kingdom', 'Germany', 'France'])]
```

```python
plt.figure(figsize=(10, 6))
plt.barh(country_summary['CountryName'], country_summary['SalePrice'], color='skyblue', label='Total Sales')
plt.barh(country_summary['CountryName'], country_summary['TotalCost'], color='orange', label='Total Costs')
plt.xlabel('Amount in USD')
plt.title('Total Sales and Total Costs by Country')
plt.legend()
plt.show()
```
