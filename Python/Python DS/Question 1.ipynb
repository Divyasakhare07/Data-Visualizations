```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('2018 Data-1.xlsx')
```

```python
unemployment_rate = data['Unemployment Rate']
obesity_rate = data['Obesity Rates']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(unemployment_rate, obesity_rate, color='yellow', marker='*', s=100, edgecolor='black')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Obesity Rate (%)')
plt.title('Unemployment Rate vs Obesity Rate (2018)')
plt.show()
```
